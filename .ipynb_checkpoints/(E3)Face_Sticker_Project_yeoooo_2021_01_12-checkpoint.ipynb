```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import dlib
```

## 처음보는 라이브러리들..
cv2와 dlib은 지금까지 접해보지 못한 라이브러리였다.  
그래서 간략하게나마 찾아봤는데, cv2, dlib 모두 이미지 처리 및 기계 학습, 얼굴인식 등을 할 수 있는 c++ 로 개발된 고성능의 라이브러리지만 <span style ="color : blue">cv2는 이미지 처리</span>에서, <span style ="color : blue">dlib는 얼굴인식</span>에서  
좀더 강점을 보인다고 한다.

## 1. 이미지 준비

```python
image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/image.jpg'
#os라이브러리를 통해 먼저 원하는 이미지의 path를 추출해준다.
img_bgr = cv2.imread(image_path)#opencv 라이브러리를 이용해 변수에 이미지 할당
                                #opencv는 bgr 형식을 기본으로 사용한다

img_bgr = cv2.resize(img_bgr,(640,360))#이미지를 640x360으로 resize
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)#bgr을 rgb로 만들어줘야한다.
img_show = img_bgr.copy()#이미지 출력을 위한 부분을 위해 따로 만들어둔다.
plt.imshow(img_bgr)
plt.show()
```

## 근데 왜 bgr..?  
bgr 형식의 이미지는 요즘 찾아보기 드물다.  
cv2 라이브러리를 이용한다면 당연히 자연스럽게 해주기 위해 rgb로 변환해줘야 하는데  
왜 bgr일까 ? 하고 찾아보았더니, 창시자가 초창기에 카메라 산업 개발자들이 bgr을  
많이 썼기 때문에 <strong><span style = "font-size : 16px"> 그냥 생각없이</span></strong> 그랬다고 한다...


```python
plt.imshow(img_rgb)#rgb 형식의 이미지 확인
plt.show()
```

# 이제 얼굴을 인식할 차례!

```python
detector_hog = dlib.get_frontal_face_detector()#호그_디텍터를 사용하기 위해 변수지정
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)
```

```python
print(dlib_rects)   # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects: #찾은 얼굴을 사각형으로 표시한다!
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)
    #사각형을 그려주는 cv2의 함수

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)

plt.imshow(img_show_rgb)
plt.show()
```

## 얼굴 인식 완료!  
하지만 이제 정확한 이목구비의 위치들을 파악해야한다.  
따라서 landmark를 표현해줄 인식 모델을 사용한다.

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)#인식 모델을 불러와 변수에 넣었다! 
```

```python
list_landmarks = []#landmark들을 넣을 list를 만들었다.
for dlib_rect in dlib_rects:#이 랜드마크에서 이미지에 대한 랜드마크 좌표값을 넣어준다!
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))#랜드마크를 총 68개 찍어준다.
```

```python
for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # nose center index : 30
    x = landmark[30][0] #코가 정중앙에 있으니 x값에 코의 landmark값을 이용했다.
    y = landmark[36][1] #코에 수염을 달 것이라서, 비교적 가까운 landmark값을 이용했다.
    #코의 x좌표와 y좌표를 받은 것
    #y값에서 덜 빼주면 덜 빼줄수록 이미지는 위로 올라간다
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
    #x, y는 코 랜드마크의 좌표
    #w, h는 얼굴 인식 box의 크기이다
    
```

## 얼굴에 대한 이미지처리 완료.  
  
  이제는 sticker에 대한 이미지처리를 해야한다

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h)) #길이와 높이를 각 108로 
print (img_sticker.shape)
#스티커 이미지를 불러오고..
```

```python
refined_x = x - w // 2  # left
#새로만들어줄 좌표 x는 코의 x축 좌표 - 얼굴 인식 box 넓이/2 
refined_y = y - h    # top
#새로만들어줄 좌표 y는 코의 y축 좌표 - 얼굴 인식 box높이
print(h)
print(y)
print(refined_y)
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
img_sticker = img_sticker[-refined_y:]
#img_sticker에 이미지 밖으로 돌출되지 않은 부분만큼만 넣어준다
print (img_sticker.shape)
```

```python
refined_y = 0 #이 때는 그냥 0으로 초기화해주나보다.. 하고 생각했다.
                #다시 생각해보니 이미지의 맨 윗부분을 지정해준 것이었다.
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
img_bgr2 = img_bgr.copy() #혹시모르니 이미지를 따로 만들어두고..
sticker_area = img_bgr[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y: img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

## 문제 발생 1.  
  
  수염이 눈썹아래로 내려오지를 않는다..  
  처음엔 img_sticker의 y좌표값을 변경시켜봤는데 눈썹 아래로는 움직일 생각을 하지 않았다.  
  face detection box의 딱 윗 부분에 걸쳐져 있는 것으로 보아 sticker가 detection box안으로 들어가지  
  못하는 것 같다고 생각했다.  

정말 많이 고민해본 뒤, 이상한 점을 발견했다. refined_y를 0으로 초기화해놓고, 그 값을 사용하지도  
않았던 것이다.  
y 값에는 더하면 더할수록 아래로 내려가니까..

```python
img_bgr3 = img_bgr2.copy()#이미지가 뒤바껴 나와 하나 더 만들었다..
refined_y =  landmark[18][1]#y값을 수정해주고
#위에 살려둔 백업 이미지 이용!
sticker_area = img_bgr2[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr2[refined_y:refined_y + img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

원래 <span style = "color : red">refined_y:img_sticker.shape[0]</span>이었던 것에
<span style = "color : blue">refined_ y</span>를 더해줬다!

```python
plt.imshow(cv2.cvtColor(img_bgr2, cv2.COLOR_BGR2RGB))
plt.show()
```

무사히 주커버그의 얼굴에 고양이 수염을 달았다..

다른 y값을 조정하려니 아래와 같은 에러가 났다.  
operands could not be broadcast together with shapes (68, 108, 3) (0, 108, 3)  
(68, 108, 3) (0, 108, 3)의 shape를 가진 피연산자는 서로 broadcast 될 수 없다는 뜻이다.
근데,  
## broadcast ?  
알기로는 broadcast는 방송의 의미를 가지고 있는데 이곳에서 쓰일 의미는 절대 아니라고 생각했다. Numpy에서 모양이 다른 배열끼리는 연산이 불가능한데, <strong><span style = "color : blue;">차원의 크기가 1(최소 1개의 배열)</span></strong>이거나 <strong><span style = "color : blue">차원의 짝이 맞을 때(=차원에 대한 축의 길이가 동일하다!)</span> </strong> 
라는 조건이 맞아진다면 모양이 다른 배열끼리도 모양을 맞춰(부족한 부분은 확장한다) 연산을 수행할 수 있다고 한다. 사실 이해가 부족해서 코드를 수정했을때 무슨 이유로 에러가 일어났는지 정확하게는 파악하지 못했다..

## 문제 발생 2.  
이번엔 계속 거슬리던 흰색 바탕과 투명 수염을 해결해야 한다.  

```python
sticker_area = img_bgr3[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr3[refined_y:refined_y + img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
```

<span style = "color : blue" >img_sticker == 0</span> 부분을 <span style ="color:blue">img_sticker == 255</span>으로 변경해주어 해결했다!  

아마 0(검은색)을 transparent로 표현 하고있었는데, 255(흰색)을 투명하게 끔 지정해주어서 그런게 아닌가 싶다.. 

```python
plt.imshow(cv2.cvtColor(img_bgr3, cv2.COLOR_BGR2RGB))
plt.show()
```
