# 연예인 닮은 꼴 찾기 프로젝트

## importing & Data preprocessing  

<strong>"원래는 셀카를 하려고 했었다 - "</strong>  
하지만 이미지를 넣어 시도하려고 할 때 마다 out of index에러가 뜨는 바람에  
셀카로 하지는 못하고, 오류가 일어나지 않는 이미지 내에서 시도해보았다.

먼저 필요한 것은 필요한 모듈과 데이터 전처리이다.  
이곳에서는 face_recognition을 통해 얼굴 인식을하고, 또 그 얼굴을 변수에 담아준다.

```python
import face_recognition #얼굴 인식을 위한 모듈
import os
import numpy as np
import dlib

image_path = os.getenv('HOME')+'/aiffel/face_embedding/images/obama.jpg'
image = face_recognition.load_image_file(image_path)
face_locations = face_recognition.face_locations(image)

print(face_locations)  # 이미지에서 얼굴 영역의 좌표

%matplotlib inline
import matplotlib.pyplot as plt

a, b, c, d = face_locations[0]
cropped_face = image[a:c,d:b,:]

plt.imshow(cropped_face)#얼굴 부분
```

```python

#얼굴 부분만 잘라내는 함수
def get_gropped_face(image_file):
    image = face_recognition.load_image_file(image_file)
    face_locations = face_recognition.face_locations(image)
    if len(face_locations) > 0:
        a, b, c, d = face_locations[0]
        cropped_face = image[a:c,d:b,:]
        return cropped_face
    else:
        return []
```

닮은 꼴을 찾아주기 위해서 대상 폴더에 들어있는 이미지를 출력해보았다.

```python
image_path = os.getenv('HOME')+'/aiffel/face_embedding/images/trump.jpg'

cropped_face = get_gropped_face(image_path)
plt.imshow(cropped_face)
```

이미지는 잘 출력되고, 이번에는 전체 파일을 확인해보았다.

```python
dir_path = os.getenv('HOME')+'/aiffel/face_embedding/images'
file_list = os.listdir(dir_path)

print ("file_list: {}".format(file_list))
```

이미지들이 잘 들어가 있는 것을 확인하였으니, 이미지를 컴퓨터가 판단할 수 있게 끔  
이미지의 임베딩 벡터를 구하는 함수를 구현해 준다

```python
image_file = os.path.join(dir_path, 'obama.jpg')#앞 인자와 두 인자를 병합하여 새 경로 생성
face = get_gropped_face(image_file)

# 얼굴 영역을 가지고 얼굴 임베딩 벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)  
embedding
```

아래 함수는 위에서 구현해 주었던 함수를 통해서 원하는 디렉토리 안에서의 이미지들을  
한꺼번에 전처리해주는 함수이다 이 함수는 크게
1. <strong>얼굴 추출</strong>
2. <strong>얼굴 이미지 임베딩 벡터 추출</strong>  
로 구성되어 있다

```python
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}

    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_gropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0:  
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]

    return embedding_dict
```

다음은 딕셔너리에 넣어준 임베딩벡터가 의도한 대로 들어가 있는지 확인한 코드이다.

```python
embedding_dict = get_face_embedding_dict(dir_path)
embedding_dict['trump']
```

무사히 임베팅 벡터화가 된 것을 확인 했으니,  
정말 닮은 꼴을 비교하기 위해서 벡터간의 거리를 확인해 줄 수 있다. 이 때 거리를 측정하기 위해 L1 distance와 L2 distance를 사용할 수 있다.

```python
def get_distance(name1, name2):
    return np.linalg.norm(embedding_dict[name1]-embedding_dict[name2], ord=2)#ord = 1 로도 쓰일 수 있다 이 둘의 차이는 L1이냐 L2이냐의 차이

get_distance('obama', 'trump')
```

obama와 trump는 0.88 정도의 거리를 가지고 있는 것으로 나왔다.

```python
# name1과 name2의 거리를 비교하는 함수를 생성하되, name1은 미리 지정하고, name2는 호출시에 인자로 받도록 합니다.
def get_sort_key_func(name1):
    def get_distance_from_name1(name2):
        return get_distance(name1, name2)
    return get_distance_from_name1

sort_key_func = get_sort_key_func('trump')   
```

```python
sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))
```

```python
def get_nearest_face(name, top=5):
    sort_key_func = get_sort_key_func(name)
    sorted_faces = sorted(embedding_dict.items(), key=lambda x:sort_key_func(x[0]))

    for i in range(top+1):
        if i == 0 : 
            continue
        if sorted_faces[i]:
            print('순위 {} : {}, {}'.format(i, sorted_faces[i][0], sort_key_func(sorted_faces[i][0])*100))
            image_path = os.getenv('HOME')+'/aiffel/face_embedding/images/'
            img = image_path+sorted_faces[i][0]+".jpg"
            img = plt.imread(img)
            plt.figure()
            plt.imshow(img)
```

시각화 단계에서는 아쉬웠다. 아직 다른 모듈을 다루는데에 익숙하지 못해 matplot을 사용했는데, 여러 이미지를 표시하기 위해 figure를 사용했다.  
하지만 한번에 여러 이미지를 출력하기 때문에 아래와 같이 순위와 함께 이미지가 나온 것이 아닌 순위가 나오고 이미지가 나오는 식이었다.

```python
get_nearest_face('박성광')
```

마지막으로 '박성광'을 nearest face의 인자로 사용했을때 비교적 낮은 수치(=비슷한 얼굴의 연예인, 0.48)을 얻어낼 수 있었다.
